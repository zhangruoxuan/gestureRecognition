```python
import numpy as np
import cv2
import os
from scipy.signal import argrelextrema
from PIL import Image
# import numpy as np
import matplotlib.pyplot as plt
```

```python
def smooth(x, window_len=13, window='hanning'):
    s = np.r_[2 * x[0] - x[window_len:1:-1],
              x, 2 * x[-1] - x[-1:-window_len:-1]]
    #print(len(s))
 
    if window == 'flat':  # moving average
        w = np.ones(window_len, 'd')
    else:
        w = getattr(np, window)(window_len)
    y = np.convolve(w / w.sum(), s, mode='same')
    return y[window_len - 1:-window_len + 1]

j=0
for videopath in vlist:
    j+=1
    cap = cv2.VideoCapture(str('/Users/zhangdaxia/Desktop/大三/数字图像处理/手语视频1/009/'+videopath))
    curr_frame = None#当前帧
    prev_frame = None#上一针
    len_window=int(10)
    frame_diffs = []#视频帧做差后的数组
    frames = []#存当前帧的
    ret, frame = cap.read()#读第一针
    i = 1
    while(ret):
        luv = cv2.cvtColor(frame, cv2.COLOR_BGR2LUV)#bgr转luv
        curr_frame = luv
        if curr_frame is not None and prev_frame is not None:
        #logic here
            diff = cv2.absdiff(curr_frame, prev_frame)#当前帧-上一针
            count = np.sum(diff)#求和
            frame_diffs.append(count)#存储差数组
            frames.append(frame)#存储当前帧
        prev_frame = curr_frame
        i = i + 1 #第几针
        ret, frame = cap.read()
    cap.release()
    imgdir ='/Users/zhangdaxia/Desktop/大三/数字图像处理/数据集/关键帧2/'+str(j)+'/'
    diff_array = np.array(frame_diffs)#转成np.array形式
    sm_diff_array = smooth(diff_array, len_window)
    frame_indexes = np.asarray(argrelextrema(sm_diff_array, np.greater))[0]#找到关键帧在数组下标
    k=0
    for i in frame_indexes:
        k+=1
        name = str(k) + ".jpg"
```

```python
import cv2
e=[]#测试
imagenum=0#存的时候名字
for n in range(1,302):
    disad='/Users/zhangdaxia/Desktop/大三/数字图像处理/数据集/关键帧2/'+str(n)
    vlist=os.listdir(disad)
    number=len(vlist)
    ad1=disad+'/1.jpg'#第一张是基准
    images=[]
    images.append(ad1)
    for num in range(2,number-1):#把每个文件夹剩下的都存images
        imagead=disad+'/'+str(num+1)+'.jpg'#地址
        images.append(imagead)
    
    raw=[]#皮肤检测后的图片
    p=0
    for i in images:#对每一个图片皮肤检测
        imagead=i
        if type(i)!=str or i=='':
            p=1
            break 
        img = cv2.imread(i, cv2.IMREAD_COLOR)#读图片
        #皮肤检测
        ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCR_CB)#
        (y, cr, cb) = cv2.split(ycrcb)
        cr1 = cv2.GaussianBlur(cr, (5, 5), 0)
        _, skin = cv2.threshold(cr1,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
        res=cv2.bitwise_and(img,img,mask=skin)
        kernel = np.ones((5,5),np.uint8)
        erode = cv2.erode(res,kernel)#腐蚀
        res=cv2.dilate(erode,kernel)
        res=cv2.cvtColor(res,cv2.COLOR_BGR2GRAY)
        raw.append(res)#存储皮肤检测后的图片
    if p==1:
        continue#防止报错
    for n in range(1,len(raw)):
        gray_diff=cv2.absdiff(raw[0],raw[n])#2-图片减去第一张
        _,thresh1 = cv2.threshold(gray_diff,40,255,cv2.THRESH_BINARY)#二值，大于40的为255，小于0
        _,thresh2 =cv2.threshold(raw[n],40,255,cv2.THRESH_BINARY)#吧关键帧转成二值
        binary =cv2.bitwise_and(thresh1,thresh2)#与运算
        kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
        erode = cv2.erode(binary,kernel)#腐蚀
        erode = cv2.erode(erode,kernel)#腐蚀
        erode = cv2.erode(erode,kernel)#腐蚀
        imagen=cv2.GaussianBlur(erode,(5,5),0)#高斯滤波
        e.append(imagen)
        x, y, w, h = cv2.boundingRect(imagen)#找到坐标
        image=cv2.imread(images[n])
        image=image[y-10:y+h,x:x+w]#截取
        imagenum+=1
        imgroad='/Users/zhangdaxia/Desktop/大三/数字图像处理/数据集/2/'+str(imagenum)+'.jpg'
        if image.size==0:
            continue
        if w*h>65000:
            continue
        if w*h<2000:
            continue
        if w<50:
            continue
        cv2.imwrite(imgroad,image)
        print(w*h)    
```
